```python
import numpy as np
import pandas as pd
```

```python
# Another index example - Air BNB dataset from Kaggle
# Default index
df_airbnb = pd.read_csv("AB_NYC_2019.csv")
df_airbnb.head(3)                        
```

```python
# Make ID column the index
df2 = df_airbnb.set_index("id")
df2.head(3)
```

```python
# Get the name for ID
id = 3647
df2.name[id]
```

```python
# Set pandas display options to all columns
numeric_columns = df2.select_dtypes(include=['float64', 'int64'])
```

```python
# Include 'room_type' column in numeric_columns DataFrame
numeric_columns['room_type'] = df2['room_type']
df3 = numeric_columns.groupby("room_type").mean()
# df3
```

```python
pd.set_option('display.max_columns', None)
df3.head(3)
```

```python
df3.index
```

```python
# Note that 'room_type' has become an index - we can again convert it into a normal column
df3 = df3.reset_index()

pd.reset_option('display.max_columns')

df3
```
