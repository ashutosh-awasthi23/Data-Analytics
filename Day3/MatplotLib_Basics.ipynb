```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate x-axis values (1 to 9)
x = np.arange(1, 10)
```

```python
# Generate y-axis values (1 to 18)
y = np.arange(1, 19)
```

```python
# Select the first 9 elements from y to match x-axis length
# Slicing to get the elements from index 0 (inclusive) to 9 (exclusive)
y = y[:9]
```

```python
# Create the line plot
plt.plot(x, y)
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('Plot Title Here')

# Increase marker size to 50 using 's' attribute
plt.scatter(x,y, s=50)


# Exporting a plot, Always export before plt.show()
plt.savefig('example.jpg')
plt.show()
```
