1. Write a Python programming to display a bar chart of the popularity of programming Languages.


Sample data:

Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


prog_lang = ['Java', 'Pyhton', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.figure(figsize = (12,8))
plt.bar(prog_lang, popularity)
plt.title("Polpularity of Programming Language Worldwide, Oct 2017 compared to a year ago")

plt.xlabel("Languages")
plt.ylabel("Popularity")

plt.grid(which='minor', linestyle='--', linewidth=0.5, color='black', alpha = 0.5) 
plt.grid(which='major', linestyle='-', linewidth=1, color='red', alpha = 0.5)
plt.minorticks_on()

plt.show()
```

10. Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.

```python
import matplotlib.pyplot as plt

# Data
men = [22, 30, 35, 35, 26]
women = [25, 32, 30, 38, 29]

# Create a figure and axes
fig, ax = plt.subplots()

# Set the width of the bars
bar_width = 0.3

# Calculate the positions for the bars
men_positions = [i - bar_width/2 for i in range(len(men))]
women_positions = [i + bar_width/2 for i in range(len(women))]

# Create bars for men
ax.bar(men_positions, men, width=bar_width, label='Men', color = 'green' )

# Create bars for women
ax.bar(women_positions, women, width=bar_width, label='Women', color = 'red')

# Customize the chart
plt.title('Scores by Person')
plt.xlabel('Person')
plt.ylabel('Scores')
plt.xticks(range(len(men)), ['G1', 'G2', 'G3', 'G4', 'G5'])
plt.legend()

# Show the plot
plt.show()
```

12. Write a Python program to create bar plots with error bars on the same figure.

```python
velocity = [0.2474, 0.1235, 0.1737, 0.1824]
sd_velocity = [0.03314, 0.02278, 0.02836, 0.02645]
x = np.arange(0, len(velocity))
plt.figure(figsize=(8,6))

plt.bar(x, velocity, yerr = sd_velocity, alpha = 0.7,
        width = 0.4, align = 'edge')

plt.title('Scores by velocity')
plt.xlabel('Velocity')
plt.ylabel('Scores')

plt.grid(which='major', linestyle='-', linewidth=0.5, color='red', alpha = 0.5)
plt.grid(which='minor', linestyle='--', linewidth=0.5, color='green', alpha = 0.5)
plt.minorticks_on()

plt.xlim(0)

plt.show()
```

13. Write a Python program to create bar plots with errorbars on the same figure. Attach a text label above each bar displaying men means (integer value).

>Sample Data:

>>Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824

>>Standard deviation of velocity: 0.03314, 0.02278, 0.02836, 0.02645


```python
velocity = [0.2474, 0.1235, 0.1737, 0.1824]
sd_velocity = [0.03314, 0.02278, 0.02836, 0.02645]

x = np.arange(0, len(velocity))
plt.figure(figsize=(8,6))

bars = plt.bar(x, velocity, yerr = sd_velocity, alpha = 1,
        width = 0.4, align = 'edge', label = "Men", color = "red",
        edgecolor = "black")

plt.title('Scores by velocity')
plt.xlabel('Velocity')
plt.ylabel('Scores')


plt.legend()
plt.xlim(0)


for bar in bars:
    yval = bar.get_height()
    x_edge = bar.get_x()
    x_width_center = bar.get_width()/2
    
    x_coordinate = x_edge + x_width_center
    
    plt.text(x_coordinate, yval + max(sd_velocity),
             f"{yval:.4f}", ha='center', va='bottom')



plt.show()
```

14. Write a Python program to create a stacked bar plot with error bars.
Note: Use bottom to stack the women?s bars on top of the men? s bars.

`Sample Data:`

`Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29)
Men Standard deviation = (4, 3, 4, 1, 5)
Women Standard deviation = (3, 5, 2, 3, 3)`




```python
plt.figure(figsize=(8,6))
plt.title("Scores by group and gender")
plt.xlabel("Groups")
plt.ylabel("Scores")

men = [22, 30, 35, 35, 26]
men_sd = [4, 3, 4, 1, 5]

women = [25, 32, 30, 35, 29]
women_sd = [3, 5, 2, 3, 3]


groups = ["Group 1", "Group 2", "Group 3", "Group 4", "Group 5"]

plt.bar(groups, men, color="red", label = "Men", yerr = men_sd, edgecolor = "black", capsize=4)

plt.bar(groups, women, color="green", label = "Women", yerr = women_sd, edgecolor = "black", bottom = men, capsize=4)



plt.legend()

plt.show()
```
