14. Write a Pandas program to create a plot of adjusted closing prices, thirty days and forty days simple moving average of Alphabet Inc. between two specific dates.


```python
df3 = df.iloc[:100, : ].copy()
df3['SMA_30_days'] = df3['Adj Close'].rolling(window = 30).mean()
df3['SMA_40_days'] = df3['Adj Close'].rolling(window = 40).mean()

df3.plot(kind = "line", x = "Date", y = ['SMA_30_days', 'SMA_40_days'], figsize=(12, 8))
plt.ylabel("Moving Average")
plt.xticks(rotation = 45)
plt.grid(True)
plt.show()
```

15. Write a Pandas program to create a plot of adjusted closing prices, 30 days simple moving average and exponential moving average of Alphabet Inc. between two specific dates.


```python
df3 = df.iloc[:100, : ].copy()
df3['SMA_30_days'] = df3['Adj Close'].rolling(window = 30).mean()
df3['EXP_30_days'] = df3['Adj Close'].ewm(span = 30).mean()

df3.plot(kind = "line", x = "Date", y = ['SMA_30_days', 'EXP_30_days'], figsize=(12, 8))
plt.ylabel("Moving Average")
plt.xticks(rotation = 45)
plt.grid(True)
plt.show()
```

17. Write a Pandas program to create a plot to visualize daily percentage returns of Alphabet Inc. stock price between two specific dates.


```python
df2['Daily Return'] = (df2['Close'] - df2["Open"])/df2["Open"] * 100

colors = ['green' if x > 0 else 'red' for x in df2['Daily Return']]

plt.figure(figsize = (12,8))
plt.bar(df2["Date"], df2["Daily Return"], width = 0.5, color = colors)
plt.xlabel("Date")
plt.ylabel("Daily Return")
plt.xticks(rotation = 45)
plt.show()
```

18. Write a Pandas program to plot the volatility over a period of time of Alphabet Inc. stock price between two specific dates.

```python
df3['Return'] = df3['Adj Close'].pct_change()

# Calculate rolling volatility (standard deviation of returns)
window_size = 30  # 30 days
df3['Volatility'] = df3['Return'].rolling(window=window_size).std() * 100  # in percentage

plt.figure(figsize=(8, 6))

plt.plot(df3['Volatility'], label='30-Day Rolling Volatility', color='blue')

plt.title('Volatility of Alphabet Inc. (GOOGL) Stock Price')
plt.xlabel('Date')
plt.ylabel('Volatility (%)')

plt.legend()
plt.grid()

plt.axhline(0, color='black', linewidth=0.8)  # Horizontal line at y=0

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
