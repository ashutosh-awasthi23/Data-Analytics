```python
# 6. Write a Python program to plot two or more lines with legends, different
# widths and colors.
# The code snippet gives the output shown in the following screenshot:


# Data
x = [10, 20, 30]
y = [20, 40, 10]
z = [10, 20, 30]
a = [40, 10, 30]

# Title and labels
plt.title("Two or More Lines on Same Plot with Suitable Legends")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.xlim(10, 30)
plt.ylim(10, 40)

line1, = plt.plot(x, y, lw=3, label='line-width 3')  # Add label for the legend
#applying comma after line1 to unpack the returning list 
#plot will return object of lines and we mainly use it in legends

line2, = plt.plot(z, a, lw=5,  label='line-width 5')  # Add label for the legend


plt.legend(handles=[line1, line2], title="Lines", loc='upper right')

# Show the plot
plt.grid()
plt.show()
```

```python
# ##7. Write a Python program to plot two or more lines with different styles.
# The code snippet gives the output shown in the following screenshot
# Data
x = [10, 20, 30]
y = [20, 40, 10]
z = [10, 20, 30]
a = [40, 10, 30]

# Title and labels
plt.title("Two or More Lines on Same Plot with Suitable Legends")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.xlim(10, 30)
plt.ylim(10, 40)

line1, = plt.plot(x, y, lw=3,ls="-",label='line-dashed')  # Add label for the legend
#applying comma after line1 to unpack the returning list 
#plot will return object of lines and we mainly use it in legends

line2, = plt.plot(z, a, lw=5,ls="-." ,label='line-dotted')  # Add label for the legend


plt.legend(handles=[line1, line2], title="Lines", loc='upper right')

# Show the plot
plt.grid()
plt.show()
```
